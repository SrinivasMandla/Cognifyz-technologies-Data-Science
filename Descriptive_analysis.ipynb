```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python

file_path = '/content/Dataset .csv'
df = pd.read_csv("/content/Dataset .csv")
```

```python
df.describe()
```

```python
print(df.columns)
df[["Average Cost for two", "Price range", "Aggregate rating", "Votes"]].describe()
```

*Explore the distribution of country code*

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Country Code', data=df, palette='cividis')
plt.title('Distribution of Country Codes')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()
```

Majority of restaurents located in country code 1




*Now explore the countrie with high no of restaurents*





```python
top_countries = df['Country Code'].value_counts().head(5)
print('top 5 countries with highest number of restaurents')
print(top_countries)
```

*Explore the distribution of the city*

```python
plt.figure(figsize=(10,8))
sns.countplot(x='City', data=df,order=df['City'].value_counts().head(20).index,palette='Set2')
plt.title('Distribution of Cities')
plt.xlabel('City')
plt.ylabel('Numberof Restaurents')
plt.xticks(rotation=90)
plt.show()
```

*Explore Top 20 Cuisines with highest number of restaurents*

```python
plt.figure(figsize=(17, 6))
cuisines_count = df['Cuisines'].value_counts()
cuisines_count.head(20).plot(kind='bar',color=sns.color_palette('Set2'))

plt.title('Top 20 Cuisines with highest number of restaurents')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.show()
```

*Top cusines and cities*

```python
top_cities = df['City'].value_counts().head(10)
print('Top 10 cities with highest number of restaurents')
print(top_cities)
```
