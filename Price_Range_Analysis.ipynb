```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)
%matplotlib inline
```

```python
most_common=df["Price range"].mode()[0]
print('most common price among all the restaurents :',most_common)
```

*calculate the average rating for each price range*

```python
Avg_rating_price=df.groupby('Price range')['Aggregate rating'].mean()
print(Avg_rating_price)
```

```python
highest_avg_rating = Avg_rating_price.idxmax()
plt.bar(Avg_rating_price.index, Avg_rating_price, color='skyblue', width=0.5) # Changed 'colors' to 'color'
plt.bar(highest_avg_rating, Avg_rating_price[highest_avg_rating], color='red', label='Highest Average Rating')
plt.xticks(rotation=0)
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.title('Average Rating by Price Range')
plt.show()

```
